```python
import pandas as pd
import os

# Load the dataset
df = pd.read_csv("statewise_jobseeker_and_vacancies_with_percentage.csv")

# Create output folder
os.makedirs("statewise_split", exist_ok=True)

# Create one CSV per state
for _, row in df.iterrows():
    state = row["State"]
    out_df = pd.DataFrame([row])
    
    # Clean filename
    safe_state = state.replace("/", "_").replace(" ", "_")
    
    out_df.to_csv(f"statewise_split/{safe_state}.csv", index=False)

print("State-wise CSVs generated in folder: statewise_split/")
```
