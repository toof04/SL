```python
import pandas as pd

# Load file
df = pd.read_csv("State_wise_Jobseeker_Registration_on_NCS_portal_till_30_June_2022.csv")

# Drop unwanted columns
df = df.drop(columns=["Male", "Female", "Transgender", "Data Upto"])

# Save cleaned CSV
df.to_csv("statewise_jobseeker_cleaned.csv", index=False)

print("Generated: statewise_jobseeker_cleaned.csv")
```

```python
import pandas as pd

# Load your previously cleaned jobseeker file
job = pd.read_csv("statewise_jobseeker_cleaned.csv")

# Load the new vacancies file
vac = pd.read_csv("State_Wise_Vacancies_Mobilised_till_30_September_2023.csv")

# Drop Data Upto from vacancy CSV
vac = vac.drop(columns=["Data Upto"])

# Standardize column names for merging
job = job.rename(columns={"States": "State"})
vac = vac.rename(columns={"States": "State"})

# Merge on State name
merged = pd.merge(job, vac, on="State", how="outer")

# Save merged file
merged.to_csv("statewise_jobseeker_and_vacancies.csv", index=False)

print("Generated: statewise_jobseeker_and_vacancies.csv")
```

```python
import pandas as pd

# Load merged file
df = pd.read_csv("statewise_jobseeker_and_vacancies.csv")

# Drop all rows that contain any NA
df_clean = df.dropna()

# Save cleaned CSV
df_clean.to_csv("statewise_jobseeker_and_vacancies_clean.csv", index=False)

print("Generated: statewise_jobseeker_and_vacancies_clean.csv")
```

```python
import pandas as pd

# Load cleaned CSV
df = pd.read_csv("statewise_jobseeker_and_vacancies_clean.csv")

# Add percentage column
df["Percentage"] = df["Mobilised Vacancies"] / df["Grand Total"] * 100

# Save final CSV
df.to_csv("statewise_jobseeker_and_vacancies_with_percentage.csv", index=False)

print("Generated: statewise_jobseeker_and_vacancies_with_percentage.csv")
```

```python
import pandas as pd

# Load file
df = pd.read_csv("statewise_jobseeker_and_vacancies_with_percentage.csv")

# Sort
top10 = df.sort_values("Percentage", ascending=False).head(10)
bottom20 = df.sort_values("Percentage", ascending=True).head(20)

# Save outputs
top10.to_csv("top10_percentage_states.csv", index=False)
bottom20.to_csv("bottom20_percentage_states.csv", index=False)

print("Generated: top10_percentage_states.csv and bottom20_percentage_states.csv")
```
