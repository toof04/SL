```python
import pandas as pd

# Load CSV
df = pd.read_csv(
    "State_UT-wise_Number_of_Schools_by_Management_and_availability_of_Girl_Toilet_and_Functional_Girl_Toilet_during_2020-21.csv"
)

# Select ONLY the required columns
output = df[
    [
        "India/State/ UT",
        "TotalGirls' and Co-educational Schools - All Management",
        "% ofGirls' and Co-educational Schools with Girl's Toilet Facility - All management"
    ]
]

# Rename for cleaner output (optional)
output = output.rename(columns={
    "India/State/ UT": "State",
    "TotalGirls' and Co-educational Schools - All Management": "Total_Girls_and_CoEd_Schools",
    "% ofGirls' and Co-educational Schools with Girl's Toilet Facility - All management":
        "Percent_Girls_Toilet_Facility"
})

# Save final CSV
output.to_csv("state_girls_school_facility_summary.csv", index=False)

print("Generated: state_girls_school_facility_summary.csv")
```
